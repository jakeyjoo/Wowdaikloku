```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://search.naver.com/search.naver'
params = {
    'query' : '놀란라이언',
    'where' : 'post'
}
```

```python
resp = requests.get(url, params = params)
resp.status_code
```

```python
soup = BeautifulSoup(resp.content, 'html.parser')
blog_contents = soup.find('ul', class_='type01')
```

```python
blog_contents.contents
```

```python
new_blog = []
for content in blog_contents.contents:
    if not str(content).strip():
        continue
    new_blog.append(content)
new_blog
```

```python
len(new_blog)
```

```python
blog_result = []
for i in new_blog:
    new_dict = {}
    a_tag = i.find('dl').find('dt').find('a')
    b_tag = i.find('dl').find('dd').find_next('dd')
#     print(a_tag)
#     print(b_tag)
    
    new_dict['title'] = a_tag.text
    new_dict['address'] = a_tag['href']
    new_dict['text'] = b_tag.text
    blog_result.append(new_dict)
print(blog_result)
```
